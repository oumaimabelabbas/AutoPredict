```python
import pandas as pd
```

```python
data_new = pd.read_csv('../../data/scrapped/avito_scraper1.csv')
data_new.head()
```

```python
#number of duplicates
data_new.duplicated().sum()
```

```python
#number of duplicates
data_new.drop_duplicates(inplace=True)
```

```python
#name of columns
data_new.columns.to_list()
```

```python
data_new = data_new.dropna(subset=['Price'])
data_new.shape
```

```python
data_new.isna().sum()
```

```python
#remove the column condition 
data_new = data_new.drop(columns=['Condition'])
```

```python
#remove the column Sector 
data_new = data_new.drop(columns=['Sector'])
```

```python
#remove the column Equipment 
data_new = data_new.drop(columns=['Equipment'])
```

```python
#renommer les colones du dataFrame data_new
data_new.rename(columns={'Gearbox':'Boite de vitesses'},inplace=True)
data_new.rename(columns={'Price':'Prix'},inplace=True)
data_new.rename(columns={'Brand':'Marque'},inplace=True)
data_new.rename(columns={'Model':'Modèle'},inplace=True)
data_new.rename(columns={'Year':'Année-Modèle'},inplace=True)
data_new.rename(columns={'Mileage':'Kilométrage'},inplace=True)
data_new.rename(columns={'Fiscal Power':'Puissance fiscale'},inplace=True)
data_new.rename(columns={'Origin':'Origine'},inplace=True)
data_new.rename(columns={'Location':'Ville'},inplace=True)
data_new.rename(columns={'First Owner':'Première main'},inplace=True)
data_new.rename(columns={'Fuel':'Type de carburant'},inplace=True)
data_new.rename(columns={'Number of Doors':'Nombre de portes'},inplace=True)
```

```python
data_new.head()
```

```python
data_new.shape
```

```python
data_new.isna().sum()
```

```python
#fill the none values in Nombre porte with 3 if they have a model in the models where Nombreporte = 3 in our data
data_new.loc[(data_new['Modèle'].isin(data_new[data_new['Nombre de portes']==3]['Modèle'].unique())) 
                  & (data_new['Nombre de portes'].isna()),'Nombre de portes']=3
```

```python
#fill the none values in Nombre porte with 3 if they have a model in the models where Nombreporte = 3 in our data
data_new.loc[(data_new['Modèle'].isin(data_new[data_new['Nombre de portes']==5]['Modèle'].unique())) 
                  & (data_new['Nombre de portes'].isna()),'Nombre de portes']=5
```

```python
data_new['Nombre de portes'].unique()
```

```python
data_new.isna().sum()
```

```python
#remove all rows that has a weird price
data_new=data_new.drop(data_new[data_new['Prix'] < 20000].index)
```

```python
data_new.shape
```

```python
data_new = data_new.dropna(subset=['Kilométrage','Marque','Modèle','Origine','Première main','Puissance fiscale','Nombre de portes'])
```

```python
data_new.isna().sum()
```

```python
data_new.shape
```

```python
# data_new['Ville'].unique()
```

```python
# remove the cities in arabic 
data_new=data_new.drop(data_new[data_new['Ville']=='الدار البيضاء'].index)
data_new=data_new.drop(data_new[data_new['Ville']=='فاس'].index)
data_new=data_new.drop(data_new[data_new['Ville']=='المحمدية'].index)
data_new=data_new.drop(data_new[data_new['Ville']=='الناظور'].index)
data_new=data_new.drop(data_new[data_new['Ville']=='القنيطرة'].index)
data_new=data_new.drop(data_new[data_new['Ville']=='طنجة'].index)
data_new=data_new.drop(data_new[data_new['Ville']=='تمارة'].index)
data_new=data_new.drop(data_new[data_new['Ville']=='مراكش'].index)
data_new=data_new.drop(data_new[data_new['Ville']=='أكادير'].index)
data_new=data_new.drop(data_new[data_new['Ville']=='تطوان'].index)
data_new=data_new.drop(data_new[data_new['Ville']=='مكناس'].index)
data_new=data_new.drop(data_new[data_new['Ville']=='الجديدة'].index)
data_new=data_new.drop(data_new[data_new['Ville']=='الرباط'].index)
data_new=data_new.drop(data_new[data_new['Ville']=='الجديدة'].index)
```

```python
data_new['Type de carburant'].unique()
```

```python
# Convert 'yes' → 'oui' and 'no' → 'non'
data_new['Première main'] = data_new['Première main'].replace({'Yes': 'Oui', 'No': 'Non'})
```

```python
data_new['Première main']
```

```python
data_new['Type de carburant'].unique()
```

```python
# traduire petrol a essence 
data_new['Type de carburant'] = data_new['Type de carburant'].replace({'Petrol': 'Essence'})
```

```python
data_new['Type de carburant'].unique()
```

```python
data_new.shape
```

```python
#remove the CV in the column Puissange fiscale  
data_new['Puissance fiscale']=data_new['Puissance fiscale'].astype(str).str.split().apply(lambda x : x[0] if len(x[0])>0 else None)
#convert to the column to num
data_new['Puissance fiscale'] = pd.to_numeric(data_new['Puissance fiscale'], errors='coerce')
```

```python
#convert the column Kilometrage to a single value
data_new['Kilométrage'] = data_new['Kilométrage'].astype(str).str.replace(' ','').str.split('-').apply(lambda x :(int(x[0])+int(x[1]))/2 if len(x)==2 else None)
```

```python
data_new.tail()
```

```python
data_new.info()
```

```python
count = (data_new['Année-Modèle'] == '1980 ou plus ancien').sum()
print(count)
```

```python
data_new = data_new.drop(
    data_new[data_new['Année-Modèle'] == '1980 ou plus ancien'].index
)
```

```python
#supprimer les lignes qui ont la colonne puissance fiscale est null
data_new.dropna(subset=['Puissance fiscale'],inplace=True)
```

```python
data_new.isna().sum()
```

```python
data_new.reset_index(drop=True,inplace=True)
```

```python
data_new['Boite de vitesses'].value_counts()
```

```python
# traduire boite de vitesses
data_new['Boite de vitesses'] = data_new['Boite de vitesses'].replace({'Manual': 'Manuelle'})
data_new['Boite de vitesses'] = data_new['Boite de vitesses'].replace({'Automatic': 'Automatique'})
```

```python
data_new['Origine']=data_new['Origine'].replace({'WW in Morocco':'WW au Maroc'})
data_new['Origine']=data_new['Origine'].replace({'Customs-cleared car':'Dédouanée'})
data_new['Origine']=data_new['Origine'].replace({'Imported New':'Importée neuve'})
```

```python
data_new = data_new[data_new['Origine'] != 'Car not yet customs-cleared']
```

```python
data_new['Origine'].unique()
```

```python
#convert the column Année-Modèle to a num
data_new['Année-Modèle'] = pd.to_numeric(data_new['Année-Modèle'], errors='coerce').fillna(0).astype(int)
```
