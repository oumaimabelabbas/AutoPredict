```python
import pandas as pd
```

# CLEAN DATA

```python
data_scrapped = pd.read_csv('../../data/scrapped/avito_cars_scrapping.csv')
```

```python
#shape of our scrapped data
data_scrapped.shape
```

```python
data_scrapped.head()
```

```python
data_scrapped.info()
```

```python
#number of duplicates
data_scrapped.duplicated().sum()
```

```python
#drop duplicates
data_scrapped.drop_duplicates(inplace=True)
```

```python
#convert the column Année-Modèle to a num
data_scrapped['Année-Modèle']=pd.to_datetime(data_scrapped['Année-Modèle'], errors='coerce').dt.year
data_scrapped['Année-Modèle']=data_scrapped['Année-Modèle'].fillna(0).astype(int) 
```

```python
#remove the column Date_pub
data_scrapped = data_scrapped.drop(columns=['Date Pub'])
```

```python
#remove the CV in the column Puissange fiscale  
data_scrapped['Puissance fiscale']=data_scrapped['Puissance fiscale'].astype(str).str.split().apply(lambda x : x[0] if len(x[0])>0 else None)
#convert to the column to num
data_scrapped['Puissance fiscale'] = pd.to_numeric(data_scrapped['Puissance fiscale'], errors='coerce')
```

```python
#remove DH from price
data_scrapped['Prix']=data_scrapped['Prix'].astype(str).str.replace('DH','').str.replace(r'\s+','', regex=True)
#convert to the column to num
data_scrapped['Prix'] = pd.to_numeric(data_scrapped['Prix'], errors='coerce')
```

```python
#convert the column Kilometrage to a single value
data_scrapped['Kilométrage'] = data_scrapped['Kilométrage'].astype(str).str.replace(' ','').str.split('-').apply(lambda x :(int(x[0])+int(x[1]))/2 if len(x)==2 else None)
```

```python
#number of nan values in each column
data_scrapped.isna().sum()
```

```python
#name of columns
data_scrapped.columns.to_list()
```

```python
#remove the none values 
data_scrapped = data_scrapped.dropna(subset=['Année-Modèle','Boite de vitesses','Type de carburant','Kilométrage',
'Marque','Modèle','Origine','Première main','Puissance fiscale','État'])
data_scrapped.shape
```

```python
#unique values of 
data_scrapped['Type de carburant'].unique()
```

```python
#fill the none values in Nombre porte with 3 if they have a model in the models where Nombreporte = 3 in our data
data_scrapped.loc[(data_scrapped['Modèle'].isin(data_scrapped[data_scrapped['Nombre de portes']==3]['Modèle'].unique())) 
                  & (data_scrapped['Nombre de portes'].isna()),'Nombre de portes']=3
```

```python
#fill the none values in Nombre porte with 3 if they have a model in the models where Nombreporte = 5 in our data
data_scrapped.loc[(data_scrapped['Modèle'].isin(data_scrapped[data_scrapped['Nombre de portes']==5]['Modèle'].unique()))
                 & (data_scrapped['Nombre de portes'].isna()),'Nombre de portes']=5
```

```python
#remove the rows where theres left none values in nombre de portes
data_scrapped=data_scrapped.dropna(subset=['Nombre de portes'])
```

```python
data_scrapped.shape
```

```python
#see number of the rows where price <20000
data_scrapped[data_scrapped['Prix']<20000]
```

```python
#remove all rows that has a weird price
data_scrapped=data_scrapped.drop(data_scrapped[data_scrapped['Prix'] < 20000].index)
```

```python
data_scrapped.shape
```

```python
data_scrapped=data_scrapped.drop(data_scrapped[data_scrapped['Origine']=='Pas encore dédouanée'].index)
```

```python
#convert the column Puissance 
data_scrapped['Puissance fiscale']= data_scrapped['Puissance fiscale'].astype(int)
```

```python
data_scrapped['Origine'].unique()
```

```python
data_scrapped2=pd.read_csv('../../data/scrapped/data_scrapped.xls')
```

# concat two clean data scrapped

```python
#concat two dataframes clean
cars_clean = pd.concat([data_scrapped,data_scrapped2])
cars_clean.shape
```

```python
cars_clean.duplicated().sum()
```

```python
cars_clean.drop_duplicates(inplace=True)
```

```python
cars_clean.reset_index(drop=True,inplace=True)
```

```python
cars_clean.columns.to_list()
```

```python
cars_clean =cars_clean.drop(columns=['État'])
```

```python
cars_clean.head()
```

# Outliers

```python
import seaborn as sns
import matplotlib.pyplot as plt
#show the boxplot for Prix
sns.boxplot(x=cars_clean['Kilométrage'])
plt.show()
```

```python
#show how much outliers we have in Kilometrage
Q1 = cars_clean['Kilométrage'].quantile(0.25)
Q3 = cars_clean['Kilométrage'].quantile(0.75)
IQR = Q3 - Q1
lower_km = Q1 - 1.5 * IQR
upper_km = Q3 + 1.5 * IQR
cars_clean[(cars_clean['Kilométrage'] < lower_km) | (cars_clean['Kilométrage'] > upper_km)]['Kilométrage'].shape
```

```python
#show the boxplot for Prix
sns.boxplot(x=cars_clean['Prix'])
plt.show()
```

```python
#show how much outliers we have in Prix

Q1 = cars_clean['Prix'].quantile(0.25)
Q3 = cars_clean['Prix'].quantile(0.75)
IQR = Q3 - Q1
lower_prix = Q1 - 1.5 * IQR
upper_prix = Q3 + 1.5 * IQR
cars_clean[(cars_clean['Prix'] < lower_prix) | (cars_clean['Prix'] > upper_prix)]['Prix'].shape
```

```python
#drop the outliers
cars_clean = cars_clean[
    (cars_clean['Prix'] >= lower_prix) & (cars_clean['Prix'] <= upper_prix) &
    (cars_clean['Kilométrage'] >= lower_km) & (cars_clean['Kilométrage'] <= upper_km)
]
```

```python
cars_clean.shape
```

```python
#save the file
cars_clean.to_csv('../../data/scrapped/cars_clean.csv',index=False)
```
