```python
# calcul des fréquences
freq = data_new['Ville'].value_counts(normalize=True)

# on identifie les villes <1 %
rare = freq[freq < 0.01].index

# on crée une nouvelle colonne
freq = data_new['Ville'].value_counts(normalize=True)
rare = freq[freq < 0.01].index
data_new['Ville_grouped'] = data_new['Ville'].replace(rare, 'Autres')

# vérification
print("Avant : villes uniques =", data_new['Ville'].nunique())
print("Après: villes uniques =", data_new['Ville_grouped'].nunique())

# bar‑plots avant/après
import matplotlib.pyplot as plt

top10_before = data_new['Ville'].value_counts().head(10)
top10_after  = data_new['Ville_grouped'].value_counts().head(10)

plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.bar(top10_before.index, top10_before.values)
plt.xticks(rotation=90)
plt.title("Top 10 villes avant")
plt.subplot(1,2,2)
plt.bar(top10_after.index, top10_after.values)
plt.xticks(rotation=90)
plt.title("Top 10 villes après regroupement")
plt.tight_layout()
plt.show()
```
